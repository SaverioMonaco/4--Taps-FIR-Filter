```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from scipy import fftpack
from numpy import savetxt
```

## Generating the data with Python

```python
# Let's see if FFT works as intended:
# We generate data with a noisy sin function and see what frequencies there are.
# we should expect a certain dominant frequency

noise = 5
time_step = 0.2
period = 10

time_vec = np.arange(0, 20, time_step)
wave = (10 * np.sin(2 * np.pi / period * time_vec))
noisy_wave = wave + noise * np.random.randn(time_vec.size)
plt.figure(figsize=(25, 5))
plt.scatter(time_vec, noisy_wave)
plt.plot(time_vec, wave, 'c')

plt.show()
```

```python
if (-np.amin(noisy_wave)>np.amax(noisy_wave)):
    norm = np.amin(noisy_wave)
else:
    norm = np.amax(noisy_wave)
```

```python
# Now we have to save our raw data into a file named input.txt
filename = 'input.txt'
```

## Frequency of the data analysis

```python
# The FFT of the signal
sig_fft = fftpack.fft(noisy_wave)

# And the power (sig_fft is of complex dtype)
power = np.abs(sig_fft)

# The corresponding frequencies
sample_freq = fftpack.fftfreq(noisy_wave.size, d=time_step)

# Plot the FFT power
plt.figure(figsize=(6, 5))
plt.xlim(0,1)
plt.plot(sample_freq, power)
plt.xlabel('Frequency [Hz]')
plt.ylabel('power')
```

```python
# Now we have to save our array in a txt file
filename = 'input.txt'

np.savetxt(filename, noisy_wave, delimiter='\n',fmt="%i") 
```

```python
fmin = -np.amin(filtered)
fmax = np.amax(filtered)
if (fmin > fmax):
    fnorm = fmin
else:
    fnorm = fmax
```

```python
filtered = norm * filtered/fnorm
```

```python
plt.figure(figsize=(25, 5))
plt.scatter(time_vec, -filtered)
```

```python
# The FFT of the signal
f_fft = fftpack.fft(filtered)

# And the power (sig_fft is of complex dtype)
power = np.abs(f_fft)

# The corresponding frequencies
f_freq = fftpack.fftfreq(filtered.size, d=time_step)

# Plot the FFT power
plt.figure(figsize=(6, 5))
plt.xlim(0,2)
plt.plot(f_freq, power)
plt.xlabel('Frequency [Hz]')
plt.ylabel('power')
```
