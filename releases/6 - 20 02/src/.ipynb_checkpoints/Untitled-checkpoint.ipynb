```python
input_sig = []
f = open("input_vectors.txt", "w")
a = False
for i in range(200):
    if(a):
        input_sig.append(120)
        f.write('%d\n' % 120)
    if(not a):
        input_sig.append(20)
        f.write('%d\n' % 20)
    if(i % 3 == 0):
        a = not a
```
